# 1 🎯 Otimização de Campanhas Publicitárias Digitais – Empresa MYK


Este projeto foi desenvolvido com o objetivo de analisar e otimizar o desempenho das campanhas publicitárias digitais, utilizando técnicas de ciência de dados e machine learning. 


A iniciativa parte de um estudo solicitado pela empresa MYK, com foco na maximização do ROI de marketing e direcionamento estratégico de anúncios.



## 1.2 🎯 Objetivos do Projeto

Identificar os principais fatores que influenciam o comportamento de clique em anúncios.

Otimizar o direcionamento de campanhas com base em características demográficas e comportamentais.

Desenvolver um modelo preditivo para estimar a probabilidade de conversão.

Segmentar a base de usuários para personalização de estratégias de marketing.

Quantificar o impacto financeiro das otimizações propostas.

## 📊  1.3 Questões Chave de Negócio

Perfil Etário e Conversão – Qual a relação entre a idade do usuário e a probabilidade de clicar em anúncios? Existe uma idade específica onde a conversão é maximizada?

Engajamento e Conversão – Como o tempo diário gasto no site influencia as taxas de clique? Existe um ponto ótimo de engajamento para maximizar conversões?

Impacto Socioeconômico – De que forma o nível de renda média da região do consumidor afeta a probabilidade de clique em anúncios?

Comportamento Digital – Qual a relação entre o tempo médio diário de uso da internet e a probabilidade de conversão em anúncios?

Segmentação Estratégica – Como podemos usar os padrões de uso da internet e tempo no site para criar segmentos de usuários com maior probabilidade de conversão?

Otimização Multifatorial – Considerando apenas as variáveis numéricas (idade, renda, tempo no site e uso da internet), qual combinação de fatores oferece maior probabilidade de conversão?

Ranking de Variáveis – Qual das variáveis numéricas disponíveis tem maior poder preditivo para determinar cliques em anúncios?

Thresholds de Conversão – Existem valores mínimos ou máximos nas variáveis numéricas (idade, tempo de uso, renda) além dos quais a probabilidade de conversão cai significativamente?


Questão Princioal: Quais públicos têm maior probabilidade de clicar em anúncios?

```python
# Criar uma figura com dois subplots (1 linha, 2 colunas)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histograma no primeiro subplot
sns.histplot(df['Age'], bins=20, kde=True, color='purple', discrete=True, ax=axes[0])
axes[0].set_title('Distribuição das Idades (Histograma)', fontsize=14)
axes[0].set_xlabel('Idade', fontsize=12)
axes[0].set_ylabel('Frequência', fontsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Boxplot no segundo subplot (orientação vertical)
sns.boxplot(y=df['Age'], color='lightgreen', ax=axes[1], orient='v')
axes[1].set_title('Resumo da Distribuição das Idades (Boxplot)', fontsize=14)
axes[1].set_ylabel('Idade', fontsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Calcular quartis e outliers
quartiles = np.percentile(df['Age'], [25, 50, 75])  # Q1, mediana (Q2), Q3
iqr = quartiles[2] - quartiles[0]  # Intervalo interquartil
lower_bound = quartiles[0] - 1.5 * iqr  # Limite inferior para outliers
upper_bound = quartiles[2] + 1.5 * iqr  # Limite superior para outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]['Age']

# Anotar os valores dos quartis
for i, q in enumerate(quartiles):
    label = ['Q1', 'Mediana', 'Q3'][i]
    axes[1].annotate(f'{label}: {q:.1f}', xy=(0.5, q), xytext=(0.6, q),
                     arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

# Anotar os outliers
if not outliers.empty:
    for outlier in outliers:
        if outlier < lower_bound:  # Outliers abaixo do boxplot
            axes[1].annotate(f'{outlier:.1f}', xy=(0.5, outlier), xytext=(0.6, outlier - 2),
                             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10, color='red')
        else:  # Outliers acima do boxplot
            axes[1].annotate(f'{outlier:.1f}', xy=(0.5, outlier), xytext=(0.6, outlier + 2),
                             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10, color='red')

# Ajustar o layout para evitar sobreposição
plt.tight_layout()

# Mostrar os gráficos
plt.show()
```

### 2.10  Country

```python
# Total

df['Country'].describe().to_frame().T
```

```python
# Função para obter o continente de um país
def get_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return 'Unknown'


# Verificar se a coluna 'Country' existe
if 'Country' not in df.columns:
    raise ValueError("A coluna 'Country' não foi encontrada no DataFrame.")

# Adicionar uma coluna "Continent" ao DataFrame
df['Continent'] = df['Country'].apply(get_continent)

# Calcular contagens por país
country_counts = df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Count']

# Mesclar com o DataFrame original para incluir o continente
country_counts = country_counts.merge(df[['Country', 'Continent']].drop_duplicates(), on='Country', how='left')

# Selecionar os top 5 países por continente
top_countries_by_continent = (
    country_counts.groupby('Continent')
    .apply(lambda x: x.nlargest(5, 'Count'))
    .reset_index(drop=True)
)

# Configurar o gráfico combinado
plt.figure(figsize=(10, 8))
sns.barplot(
    data=top_countries_by_continent,
    y='Country',
    x='Count',
    hue='Continent',
    palette='Set2'
)

# Configurar título e rótulos dos eixos
plt.title('Top Países por Continente')
plt.xlabel('Número de Consumidores')
plt.ylabel('País')
plt.legend(title='Continente')

# Ajustar layout
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

*Interpretação:*


**Visualização dos Top 5 Países por Continente:**

O gráfico utiliza barras horizontais para representar o número de consumidores por país.

As barras são coloridas de acordo com o continente ao qual cada país pertence.

O eixo x representa o número de consumidores, variando de 0 a 8.

O eixo y lista os países em ordem decrescente de consumidores.

## 3.1 Análise Bidimensional

```python
# Definindo estilo para visualizações
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")


# 1. Pré-processamento dos Dados
# Separando variáveis numéricas e categóricas
numeric_cols = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']
categorical_cols = ['Ad Topic Line', 'City', 'Country', 'Timestamp']

# Pipeline para codificação one-hot e padronização
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols)  
        # Configuração handle_unknown='ignore' adicionada para resolver o erro:
        # "Found unknown categories [...] in column [...] during transform"
        # Isso faz com que o encoder ignore categorias desconhecidas no conjunto de teste.
    ])

# Dividindo em X e y
X = df.drop('Clicked on Ad', axis=1)
y = df['Clicked on Ad']

# Divisão em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o pipeline com regressão logística
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
])

# Treinando o modelo
model.fit(X_train, y_train)

# 2. Avaliação do Modelo
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

print("\n--- Avaliação do Modelo ---")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusão:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Visualização da matriz de confusão
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Matriz de Confusão', fontsize=16)
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.tight_layout()
plt.savefig('matriz_confusao_logistica.png')

# Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC', fontsize=16)
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('roc_curve_logistica.png')

# 3. Interpretação dos Coeficientes
# Obtendo os nomes das colunas após transformação
numeric_features = numeric_cols
categorical_features = model.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(categorical_cols)
all_features = numeric_features + list(categorical_features)

# Obtendo os coeficientes do modelo
coefficients = model.named_steps['classifier'].coef_[0]
feature_importance = pd.DataFrame({
    'Feature': all_features,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', ascending=False)

print("\n--- Coeficientes da Regressão Logística ---")
print(feature_importance)

# Visualização dos coeficientes
plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=feature_importance.iloc[:20])  # Exibindo as 20 principais
plt.title('Coeficientes da Regressão Logística', fontsize=16)
plt.xlabel('Coeficiente')
plt.ylabel('Variável')
plt.tight_layout()
plt.savefig('coeficientes_logistica.png')

# 4. Sistema de Scoring
def calculate_propensity_score(user_data):
    # user_data deve ser um DataFrame com as mesmas colunas de X
    propensity = model.predict_proba(user_data)[0, 1]
    return propensity

# Exemplo de uso do sistema de scoring
example_user = X_test.iloc[0:1]  # Pegando uma linha de teste como exemplo
score = calculate_propensity_score(example_user)
print("\n--- Sistema de Scoring - Exemplo ---")
print(f"Score: {score:.4f}")

print("\n=== FIM DA ANÁLISE ===")
```

**A Regressão Logística:**


É uma técnica de aprendizado supervisionado amplamente utilizada para problemas de classificação binária, (prever se um usuário clicará ou não em um anúncio)

**Curva ROC (Receiver Operating Characteristic)**

   *Gráfico:*

O gráfico mostra a Curva ROC , que é uma representação da relação entre a Taxa de Verdadeiros Positivos (TPR) e a Taxa de Falsos Positivos (FPR) para diferentes thresholds de classificação.

A área sob a curva (AUC) é indicada como 0.988 .          

             

 **Matriz de Confusão** 

É uma ferramenta essencial para avaliar o desempenho de um modelo de classificação. Ela mostra como o modelo classificou corretamente ou incorretamente as classes previstas em relação às classes reais.
